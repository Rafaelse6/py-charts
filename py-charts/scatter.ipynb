```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Scatter plot considering volume and scatter
plt.scatter(x=base.Girth, y=base.Volume, color='blue',
            facecolors='none', marker='*')
plt.title('Árvores')
plt.xlabel('Circunferências')
plt.ylabel('Volume')
```

```python
# Scatter plot considering volume and "girth" attribute without marker
plt.plot(base.Girth, base.Volume)
plt.title('Ávores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Scanner plot with data removal(jitter)
# fit_reg = tendency line
sns.regplot(x=base.Girth, y=base.Volume,
            data=base, x_jitter=0.3, fit_reg=False)
```
